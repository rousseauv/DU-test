### Redefine all these object with dollars replacing , with . and without dollars

```python
df['INCOME'] = df.INCOME.str.replace('$', '')
df['INCOME'] = df.INCOME.str.replace(',', '.').astype(float)
```

```python
df['HOME_VAL'] = df.HOME_VAL.str.replace('$', '')
df['HOME_VAL'] = df.HOME_VAL.str.replace(',', '.').astype(float)
```

```python
df['BLUEBOOK'] = df.BLUEBOOK.str.replace('$', '')
df['BLUEBOOK'] = df.BLUEBOOK.str.replace(',', '.').astype(float)
```

```python
df['OLDCLAIM'] = df.OLDCLAIM.str.replace('$', '')
df['OLDCLAIM'] = df.OLDCLAIM.str.replace(',', '.').astype(float)
```

## Analysis of predictor variables

### Separate num(erical) parameters and cat(egorical) parameters

```python
PARAM = ['TARGET_FLAG', 'TARGET_AMT', 'KIDSDRIV', 'AGE', 'HOMEKIDS',
       'YOJ', 'INCOME', 'PARENT1', 'HOME_VAL', 'MSTATUS', 'SEX', 'EDUCATION',
       'JOB', 'TRAVTIME', 'CAR_USE', 'BLUEBOOK', 'TIF', 'CAR_TYPE', 'RED_CAR',
       'OLDCLAIM', 'CLM_FREQ', 'REVOKED', 'MVR_PTS', 'CAR_AGE', 'URBANICITY']
PARAM_NUM = ['TARGET_AMT', 'KIDSDRIV', 'AGE', 'HOMEKIDS',
       'YOJ', 'INCOME', 'HOME_VAL', 'TRAVTIME', 'BLUEBOOK', 'TIF',
       'OLDCLAIM', 'CLM_FREQ', 'MVR_PTS', 'CAR_AGE']
PARAM_CAT = ['PARENT1', 'MSTATUS', 'SEX', 'EDUCATION',
       'JOB', 'CAR_USE', 'CAR_TYPE', 'RED_CAR', 'REVOKED', 'URBANICITY']
```

### Boxplot for numeric variables

```python
fig1, axes1 = plt.subplots(2,7)

for var, ax in zip(PARAM_NUM, axes1.flatten()):
    sns.boxenplot(y=var, x="TARGET_FLAG", data=df, ax=ax)
    
plt.tight_layout()
```

### Barplot for categorical variables

```python
fig2, axes2 = plt.subplots(2, 5)
for var, ax in zip(PARAM_CAT, axes2.flatten()):
    sns.barplot(x=var, y="TARGET_FLAG", data=df, ax=ax)
    for label in ax.get_xticklabels():
        label.set_rotation(90)
plt.tight_layout()
```

### Plot histograms for numeric variables

```python
df[PARAM].hist(bins=15, figsize=(15, 15), layout=(5, 5));
```

### Calculate and plot correlation matrix

```python
corrMatrix = df[PARAM_NUM].corr()
```

```python
sns.heatmap(corrMatrix, annot=True, cmap="PiYG", center=0)
plt.figure()
plt.show()
```
